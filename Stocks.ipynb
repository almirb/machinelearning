```python
#fetch the data from github
import os
from six.moves import urllib

DOWNLOAD_ROOT="https://github.com/CNuge/kaggle_code/blob/master/"
STOCKS_PATH=os.path.join("datasets","stocks")
STOCKS_URL=DOWNLOAD_ROOT+"stock_data/all_s_and_p_data.csv?raw=true"
STOCKS_FILENAME="stocks.csv"

def fetch_stock_data(stocks_url=STOCKS_URL,stocks_path=STOCKS_PATH):
    if not os.path.isdir(stocks_path):
        os.makedirs(stocks_path)
    file_path=os.path.join(stocks_path,STOCKS_FILENAME)
    urllib.request.urlretrieve(stocks_url,file_path)

fetch_stock_data()
```

```python
# Load the data using pandas
import pandas as pd


def load_csv_data(stocks_path=STOCKS_PATH,stocks_fileName=STOCKS_FILENAME):
    csv_path=os.path.join(stocks_path,stocks_fileName)
    return pd.read_csv(csv_path)

csv_data=load_csv_data()
csv_data.head()
```
