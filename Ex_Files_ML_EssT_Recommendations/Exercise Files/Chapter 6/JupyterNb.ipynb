```python
import numpy as np
import pandas as pd
import matrix_factorization_utilities as mfu
```

```python
# Load user ratings
user_ratings_df=pd.read_csv("movie_ratings_data_set.csv")
# Load movie titles
movies_df=pd.read_csv("movies.csv")
```

```python
# Convert the running list of user ratings into a matrix
ratings_pivot_df=user_ratings_df.pivot_table(index="user_id", columns="movie_id",aggfunc=np.max)
```

```python
U,M = mfu.low_rank_matrix_factorization(ratings_pivot_df.as_matrix(),num_features=15,regularization_amount=0.1)
```

```python
predicted_ratings=np.matmul(U,M)
```
